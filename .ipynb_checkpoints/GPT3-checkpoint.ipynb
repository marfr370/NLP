### Tokenizing the text to remove lyrics over 2000 tokens

```python
#!pip install transformers (in ipykernel)
# Had to install jupyter nbextension enable --py widgetsnbextension and !pip install ipywidgets in jupyter virenv. 


from transformers import GPT2TokenizerFast
tokenizer = GPT2TokenizerFast.from_pretrained("gpt2")

# remove reviews that are too long
subset_lyrics['n_tokens'] = subset_lyrics.Lyric.apply(lambda x: len(tokenizer.encode(x)))
subset_lyrics = subset_lyrics[subset_lyrics.n_tokens<2000].tail(10_000)
```

### Finding embedding vectors for lyrics 

```python
def get_embedding(text, engine="babbage-similarity"):
    text = text.replace("\n", " ")
    return openai.Engine(id=engine).embeddings(input = [text])['data'][0]['embedding']

subset_lyrics['babbage_similarity'] = subset_lyrics.Lyric.apply(lambda x: get_embedding(x))
```

### t-SNE visialization and then classification using random forest and xgboost

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
matrix = subset_lyrics.babbage_similarity.to_list()

# Create a t-SNE model and transform the data
tsne = TSNE(n_components=2, perplexity=15, random_state=42, init='random', learning_rate=200)
vis_dims = tsne.fit_transform(matrix)
colors = {'Rock':'tab:red', 'Pop':'tab:green', 'Hip Hop':'tab:blue', 'Sertanejo':'tab:purple', 'Samba':'tab:orange'}
#colorlist = ["red", "purple", "pink", "darkgreen", "darkorange"]
#colormap = matplotlib.colors.ListedColormap(colors)

x = [x for x,y in vis_dims]
y = [y for x,y in vis_dims]
 
plt.scatter(x, y, c=subset_lyrics['Genre'].map(colors), alpha=0.3)
plt.title("Song lyrics visualized in language using t-SNE")
```

```python
from sklearn.model_selection import train_test_split
 
X_train, X_test, y_train, y_test = train_test_split(
    list(subset_lyrics.babbage_similarity.values),
    subset_lyrics.Genre,
    test_size = 0.2,
    random_state=42
)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
 
clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train, y_train)
preds = clf.predict(X_test)
```

```python
print(classification_report(y_test, preds))
```
